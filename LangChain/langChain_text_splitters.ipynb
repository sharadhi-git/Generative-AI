```python
#laod text files form the txt folder directory
#import the directory loader
from langchain_community.document_loaders import DirectoryLoader

#initialize a directory loader with the path to the directory and a glob pattern for text fies
loader=DirectoryLoader("/Users/shara/OneDrive/Documents/Coding Stuff/Generative AI/LangChain/Documents/txt folder",glob="**/*.txt")

#load the text data from the directory
dataset=loader.load()
for data in dataset:
    print("---------------------")
    print(data.page_content)
    print(data.metadata)
```

```python
#To calculate the token count of each document, which means the chracther count of each document we use the len function
# here we caclualte the tokens based on charachters(a chrchter -level tokenizer), but there are several methods that we can clacualte the tokens,
#as an example, chat gpt uses the cl100k_base tiktoken tonkenizer method. 
tokens_count=[len(data.page_content) for data in dataset]

print(tokens_count)
```

```python
#Get the encoding for a tiktok tokenizer, we will use this tokenizer method in this application
#first we install the library
!pip install tiktoken -qU
```

```python

import tiktoken
tokenizer_model=tiktoken.encoding_for_model('gpt-3.5-turbo')
print(tokenizer_model)
```

```python
#Get the encoding for our tokenizer
tokenizer=tiktoken.get_encoding('cl100k_base')

#create a function to calculate the length of text in tokens using tiktoken
def tiktoken_len(text):
    tokens=tokenizer.encode(text)
    token_length=len(tokens)
    print(token_length)
    return token_length
```

```python
tiktoken_method_token_count=[tiktoken_len(str(data)) for data in dataset]
print(tiktoken_method_token_count)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

#intialize the text splitter with a chunck size of 50 and no overlap.
#here we use a chachter-level function to count the length of chunks(not the tiktoekn method at this moment)
text_splitter=RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=0,length_function=len)#which means we create chunks with chrhactersize of 50.
```

```python
#split the chunks for the first document
chunks=text_splitter.split_text(dataset[0].page_content)
len(chunks)

print(len(chunks))
print(chunks[1]) #print the first chunk of the first docuemnt in the dataset.
print(len(chunks[1]))# gives a toekn size which is less than maximum chunksize of 50.
```
