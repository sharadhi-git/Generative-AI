Loading text files

```python
from langchain_community.document_loaders import TextLoader

#initialize the Textloader with the path for the text file
loader=TextLoader("/Users/shara/OneDrive/Documents/Coding Stuff/Generative AI/LangChain/Documents/example_txt_file.txt")
#load the text dat
txt_data=loader.load()
print(txt_data)
```

**Loading text files from a directory.**
**For this purpose, we need unstructured pacakge**

```python
!pip install unstructured -qU
```

```python
#import the directory loader
from langchain_community.document_loaders import DirectoryLoader

#initialize a directory loader with the path to the directory and a glob pattern for text fies
loader=DirectoryLoader("/Users/shara/OneDrive/Documents/Coding Stuff/Generative AI/LangChain/Documents/txt folder",glob="**/*.txt")

#load the text data from the directory
dataset=loader.load()
for data in dataset:
    print("---------------------")
    print(data.page_content)
    print(data.metadata)
```

**Loading pdf files from a directory**

```python
from langchain_community.document_loaders import PyPDFDirectoryLoader

#initialize a directory loader with the path to the directory and a glob pattern for text fies
loader=PyPDFDirectoryLoader("/Users/shara/OneDrive/Documents/Coding Stuff/Generative AI/LangChain/Documents/pdf folder",glob="**/*.pdf")

#load the text data from the directory
dataset=loader.load()
for data in dataset:
    print("---------------------")
    print(data.page_content)
    print(data.metadata)
```

```python
from langchain_community.document_loaders import BSHTMLLoader
#initialize the pdf document loader

loader = BSHTMLLoader(
    "/Users/shara/OneDrive/Documents/Coding Stuff/Generative AI/LangChain/Documents/example_html_file.html",
    bs_kwargs={"features": "html.parser"}
)
html_data = loader.load()
print(html_data)
```
